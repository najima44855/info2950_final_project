Below are some functions to sift through the raw Spotify API data. For the first function, playlist IDs can be found in spotify and are part of the Spotify URIs. `get_track_ids_in_playlist` returns a list of track IDs for a certain playlist and are collected from a request that uses our access token from the web API. For each track ID, we can use them to get a dictionary with information about each track.

```python
def get_track_ids_in_playlist(playlist_id):
    """ Builds a list of track IDs from the playlist ID
    
    Arguments
    =========
    
    playlist_id: string.
        This can be retrieved from going to a playlist on Spotify, clicking
        'Share', then 'Copy Spotify URI'
    
    Returns
    =======
    
    track_ids: list.
        The list contains all track id's in the playlist.
    """
    track_ids = []
    
    headers = {
        'Authorization': 'Bearer {token}'.format(token=access_token)   
    }

    r = requests.get(BASE_URL + 'playlists/' + playlist_id, headers=headers)

    tracks = []   # will hold all track info

    for track in r.json()['tracks']['items']:
        #print(track['track']['name'])
        track_ids.append(track['track']['id'])
        
    return track_ids
```

```python
def get_audio_features_from_track(track_id):
    """ Returns the json of the audio features of a track.
    
    Arguments
    =========
    
    track_id: string.
        This can be retrieved from going to a playlist on Spotify, clicking
        'Share', then 'Copy Spotify URI'
    
    Returns
    =======
    
    audio_features: dict.
        The json contains all audio features in the track.
    """
    r = requests.get(BASE_URL + 'audio-features/' + track_id, headers=headers)
    return r.json()
```

**Data Cleaning and Exploratory Analysis**

Below, we clean our data by removing unnecessary columns. Since our research will focus on the comparison of music across countries, we will not include a column for each playlist that sorts the music's popularity. 

```python
songs = pd.DataFrame()
for country in spotify_ids:
    print(country)
    tracks_for_country = get_track_ids_in_playlist(spotify_ids[country])
    track_info_df = pd.DataFrame([get_audio_features_from_track(x) for x in tracks_for_country])
    track_info_df["country"] = country
    songs = songs.append(track_info_df, ignore_index = True)
    
songs = songs.drop(columns = ["type", "uri", "track_href", "analysis_url", "error"])
songs = songs.dropna() # Remove rows that the API returns errors on
songs = songs.reset_index()
```

**Data Limitations**

Since we are only looking at the top 50 songs from each country, we might not get a wholistic picture of a country's music tastes. If a country has only pop and R&B songs in their top 50 playlist, we won't get to analyze the characteristics of music of other genres like classical or jazz for a certain country. Another limitation of our data is the fact that these playlist charts are only for a single day for a country. Music tastes and music choices often change as new songs are released and our data analysis is only focused on a single day, which might not paint a complete picture of a country's tastes unless we bring in data from different time periods. Since Spotify doesn't keep old top 50 charts available to the public, it might be hard to collect old music data. One final limitation of our data is that our data only focuses on the songs themself and not on the listeners. Having information about listeners like when they listen or what ages listen to certain songs would be helpful in discerning information about music habits/preferences in a country.

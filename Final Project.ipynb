### Analysis of song positivity and governing styles of countries

```python
def create_gov_hist(data_frame, characteristic):
    """
    Creates histograms showing the frequency of characteristic for each governing style from data_frame. For 
    each histogram, the mean characteristic value is graphed by a red vertical line. Each histogram shares 
    the same x axis scale. The mean values for each governing system is also printed. 
    
    Requires: data_frame is a pandas data frame object. Must contain a"goverenment" column. Characteristic is a 
    string that is a column name in data_frame.
    """
    # Standardize x axis to be from 0 to 1     
    gov_syst_hist = data_frame[characteristic].hist(by = data_frame["government"], 
                                                    sharex = True, figsize = (20, 10)) 
    
    # Calculate mean danceability of each histogram and get sorted list of different goverenment types
    mean_dance = data_frame.groupby(by = "government").mean()
    government_cols = list(data_frame["government"].unique())
    government_cols.sort()
    row_iterator = 0
    
    for ax in gov_syst_hist.flatten():
        ax.set_xlabel(characteristic)
        ax.set_ylabel("Song frequency")
        # If we only are creating three histograms, pd.hist creates four anyways so we need to catch indexing exceptions
        try: 
            ax.axvline(mean_dance[characteristic][government_cols[row_iterator]], 0, 100, c = "red")
        except:
            pass
        row_iterator += 1
         
    pl.suptitle("Histogram of song " + characteristic + " sorted by governing systems")
    plt.show()
    print("Mean " + characteristic + " of each governing system \n \n {}"
          .format(mean_dance[characteristic].sort_values(ascending = False).round(3)))
```

We see that the highest mean dancaebility are countries with democratic, presidential systems with monarchies danceability scores being below democratic systems and more authoritarian states being the lowest. One interesting result from the mean danceability is how low the Semi Presidential mean is. The mean is lower than most of the monarchy governing style means. We thought that the democratic countries would have the highest danceability means, which is counter to what the mean table presents.

Next, we look at where the mean danceability of music in full presidential countries would fall on the distribution of song danceability in authoritarian states by calculating z scores. We will consider Military Junta and One-party states as our authoritarian states. We assume that the histograms are normally distributed as z scores won't work with data that is not normally distributed. 

```python
def calc_z(data_frame, category, gov_one, gov_two):
    """ Calculate the z score of gov_one mean based on gov_two's mean and standard deviation"""
    mean = data_frame.groupby(by = "government").mean()
    std = data_frame.groupby(by = "government").std()
    print("Deviation of mean " + gov_one + " " + category + " from " + gov_two + " " + category + " mean: {:.3f}"
          .format((mean[category][gov_one] - mean[category][gov_two]) / std[category][gov_two]))
```

### Analysis of government trust and song positivity

To begin this analysis, we import data from the research paper titled "Trust" from Our World in Data. <br>

We group the mean danceability, energy, and valence of each country together. Then we train linear regression models that input the different song characteristics and output the government trust rates. We do both single variable and multivariate linear regression models.

```python
means_by_country = songs_trust.groupby(by = "country").mean()
correlation_matrix = means_by_country.corr()

def create_lin_reg(mean_data, corr_mat, char):
    """
    Train a linear regression model that inputs char and outputs government trust. Prints the pearson
    and spearman correlation coefficient, the coefficient of determination, as well as creates a scatter plot 
    of government trust vs. characteristic with both the predicted values of the linear regression model and 
    the true values. Later on, we calculate residuals.
    
    Requires: mean_data and corr_mat are dataframes. mean_data contains mean stats grouped by each country.
    corr_mat and corr_mat_sp are the pearson and spearman correlation matrices from mean_data. char
    is a descriptor of a song (danceability, valence, etc.).
    """
    print("Indicator variable: {}".format(char))
    lr = LinearRegression().fit(mean_data[[char]], mean_data["trust"])
    corr_matrix = mean_data.corr()
    print("The slope of mean " + char + " by country vs. country government trust: {:.3f}"
          .format(lr.coef_[0]))
    print("The pearson correlation coefficient: {:.3f}"
          .format(corr_matrix.loc["trust", char]))
    print("The coefficient of determination: {:.3f}"
          .format(lr.score(mean_data[[char]], mean_data["trust"])))

    y_hat = lr.predict(mean_data[[char]])
    plt.scatter(mean_data[char], mean_data["trust"], label = "True values") 
    plt.scatter(mean_data[char], y_hat, label = "Predictions based on lr model")
    plt.legend()
    plt.title("Government trust by country vs. mean country song " + char + " colored by predicted or true values")
    plt.xlabel("Mean song " + char)
    plt.ylabel("Government trust")
    plt.show()
```

```python
create_lin_reg(means_by_country, correlation_matrix, "danceability")
```

From first glance of our scatter plot, we see a loose negative correlation between danceability and government trust. According to our linear regression model, as danceability increases by $1$, or a song goes from not danceable at all to extremely danceable, a countries trust in their government decreases by $-1.965$, or by $196$ percent. If we scale this coefficient to better fit our data, as danceability increases by $0.1$, government trust in a country decreases by $19.6$ percent. The linear relationship between danceability and government trust is  moderately correlated, so our coefficient for our linear regression model could be a good representation of the relationship between mean song danceability and government trust. ### Government trust and song positivity significance

To ensure that our linear regression models are a good representation of the relationship between song characteristic and government trust, we look at the residuals of some of the linear models as well as the p values.

```python
def calc_residuals_p_value(mean_data, char):
    """
    Graphs the residuals of the linear regression model with the values of a certain song characteristic on 
    the x axis. 
    Also calculates the p value. 
    """
    lr = LinearRegression().fit(mean_data[[char]], mean_data["trust"])
    y_hat = lr.predict(mean_data[[char]])
    chi_squared_results = chisquare(mean_data["trust"], f_exp=y_hat)
    residuals = np.subtract(mean_data["trust"], y_hat)
    
    residual_df = pd.DataFrame(residuals.tolist(), columns=["residuals"])
    residual_df[char] = mean_data[char].tolist()
    print("The p-value of government trust and y_hat: {:.3g}".
        format(chi_squared_results.pvalue))
    print("The correlation coefficient of the residual graph: {:.3g}".
        format(residual_df[char].corr(residual_df["residuals"])))
    
    plt.scatter(mean_data[char], residuals)
    plt.title("Residuals vs. observed X values")
    plt.xlabel("Observed values of " + char)
    plt.ylabel("Linear model residuals for observed values of government trust")
    plt.show()
```

```python
calc_residuals_p_value(means_by_country, "danceability")
```

```python
calc_residuals_p_value(means_by_country, "speechiness")
```

```python
calc_residuals_p_value(means_by_country, "valence")
```

We see that the residuals are random in that when the danceability, speechiness, or valence of a song increases, we don't seem to see any pattern to how the residuals change. Furthermore, the residual plot is randomly scattered. Finally, the p-values are all very high, meaning there is no significant difference between the expected values (which are linear and from the linear regression model) and our observed values. This provides good evidence that a linear model is a good representation of the relation between song positivity and government trust and that there is a negative relationship between song positivity and government trust.

Next, we train a multivariate linear regression model that takes in danceability, energy, and speechiness as inputs. We decided to leave out valence because valence's correlation coefficient values are lower than the others. We leave out acousticness because this variable is similar to speechiness, which we plan on including.

```python
multi_var_input = means_by_country[["danceability", "energy", "speechiness"]]
multi_lr = LinearRegression().fit(multi_var_input, means_by_country["trust"])
coefficients = multi_lr.coef_
print("The r squared for multi variate model: {:.3f}"
      .format(multi_lr.score(multi_var_input, means_by_country["trust"])))
print("Slope of danceability: {:.3f}".format(coefficients[0]))
print("Slope of energy: {:.3f}".format(coefficients[1]))
print("Slope of speechiness: {:.3f}".format(coefficients[2]))
```

We see that the coefficient of determination has increased in the multivariate model, leading us to believe that the multivariate model is a better predictor of government trust when inputting song characteristics. All of the coefficients have decreased in magnitude but are still negative, which strengthens our initial conclusion that sadder songs are correlated to higher government trust.

Next, we look at how well our kmeans clustering algorithms went. First, we look at the kmeans algorithm for danceability and energy. We create a lookup table for the labels from $0$ to $2$. Next, we calculate the accuracy of the kmeans algorithm on enerty and valence.

```python
lookup_ev = pd.DataFrame({"labels_ev": [0, 1, 2], "predicted_continent_ev": ["Asia", "Americas", "Europe"]})
country_mns = pd.merge(country_mns, lookup_ev, on = "labels_ev")
```

```python
country_mns.head()
```

```python
continent_accuracy = country_mns[country_mns["continent"] == country_mns["predicted_continent_ev"]]
print("The accuracy of the kmeans algorithm in predicting continent on energy and valence: {:.1%}"
      .format(continent_accuracy.shape[0]/country_mns.shape[0]))
```

Since we seem to have different accuracies for the kmeans algorithm for each continent, we want to look at the prediction accuracy by continent.

```python
def continent_accuracy(continent, df, predictor):
    """
    Calculates the prediction accuracy of a single continent
    """
    of_continent = df[df["continent"] == continent]
    rows_correct = of_continent[of_continent["continent"] == of_continent[predictor]]
    return ((rows_correct.shape[0]/of_continent.shape[0]) * 100)
```

```python
prediction_accuracy_ev = lookup_ev.copy()
prediction_accuracy_ev["prediction_percentage"] = lookup_ev["predicted_continent_ev"].apply(
    continent_accuracy, df = country_mns, predictor = "predicted_continent_ev")
prediction_accuracy_ev["prediction_percentage"] = prediction_accuracy_ev["prediction_percentage"].round(1)
```

```python
prediction_accuracy_ev
```

We see that when looking at energy and valence, the continent with the highest prediction accuracy is the Americas. Asia is nearly thirty percent lower while Europe seems to be close to the Americas. This is probably because a lot of the countries that should be labeled as Asian are actually mislabeled as Europe since the points of energy vs. valence for Asian and European points are so close to each other.

```python
prediction_accuracy_de = lookup_de.copy()
prediction_accuracy_de["prediction_percentage"] = lookup_de["predicted_continent_de"].apply(
    continent_accuracy, df = country_mns, predictor = "predicted_continent_de")
prediction_accuracy_de["prediction_percentage"] = prediction_accuracy_de["prediction_percentage"].round(1)
```

```python
prediction_accuracy_de
```

The accuracy of the Europe and Asia countries are the same compared to the energy and valence kmeans accuracy, but the American prediction accuracy is a little lower. From this, we could consider that danceability might not be as strong as a predictor of song positivity than energy or valence. However, the prediction percentage of the Americas is still in the eighty percent, which is a high accuracy so we think that danceability is still a good predictor. 

## Interpretation and conclusions

Governing system relationship to song positivity: after looking at song danceability, energy, and valence, and their relation to countries governing styles, we can discern a few different observations. First of all, countries with full presidential systems consistently score highest on all of our positivity metrics while countries with one-party states or military junta states usually score the lowest on all of our positivity metrics with a difference of around $0.15$ "points" on a $0$ to $1$ scale. One reason for why this could be is that people in more authoritarian countries tend to face more government repression which leads to more reserved and timid lifestyes. As a result, these citizens wouldn't be inclined to listen to extremely upbeat or happy music but instead of sadder or more contained songs. Countries with monarchies tend to fall in the middle of the song positivity scores but skewed towards authoritarian governing systems. One reason for why this could be is that the governing style of monarchies aren't exactly halfway between democratic and authoritarian systems but tend to follow more authoritarian policies. To check the significance of the means of different governing systems, we looked at where the means of different governing systems would fall on other governing system distributions. We concluded that when looking at song danceability, energy, and valence, about $75$ percent of the songs had higher positivity than the mean song positivity of authoritarian countries. From this, we can conclude that the difference between song positivity in democratic and authoritarian states is significant and that there is a noticeable difference in the song positivity of these two governing systems and not just due to pure variance. The only governing style outlier seems to be the semi presidential positivity scores which range from low to high when comparing its song positivity metrics to other governing styles. One reason for why this could be is that semi presidential governing systems, in reality, are not democratic but tend to be facades for more authoritarian regimes. Thus, while semi presidential systems may be labeled as "democratic", their song positivity results align more with authoritarian governing styles like one-party states. 

Governing trust relationship to song positivity: After training our linear regression models on song characteristics, we see that there is a negative linear correlation between government trust and danceability, energy, valence, and acousticness with the indicator variables danceability, speechiness, and energy being the strongest predictors of government trust. In our governing style and song positivity analysis, we saw that democratic countries had higher song positivity. In this analysis, we see that countries with higher song positivity have lower government approval ratings. These results seem to contradict each other in that we thought countries with democratic systems would have higher approval ratings and thus higher song danceability. In fact, we see the opposite. This could occur for a few reasons. First, perhaps democratic countries do not have higher government trust ratings. However, we do not think this because compared to authoritarian countries, democratic countries probably have more trust in their governments to create legislation that serves the public. On the other hand, authoritarian governments might manipulate their citizenship into believing that the government is trustworthy, raising the countries' government trust rates of authoritarian governments. Another reason this could occur is that governing style and government trust rates do not have any relation and that we must look at these two analyses separately. Lastly, the negative correlation in government trust and song positivity could be due to the Simpson's paradox. To check for the Simpson's paradox, we created a multivariate model with all four inputs, which we believed better predicted government trust since the coefficient of determination was higher. However, the coefficients were still negative with the multivariate model so we don't think the Simpson's paradox is why we see "flipped" results.

Geography relationship to song preferences: We found that there is a difference in the song listening habits if we split countries by continents. The song preferences of the Americas (North and South) tend to be more separated from the listening preferences of European and Asian countries. However, the preferences of the American countries are a lot more concentrated and tend to be more positive. This could be explained by the fact that the Americas are relatively isolated from other continents and thus developed their own specific musical tastes. This would explain why the k-means algorithm prediction accuracy for America is higher and why the song preferences are extremely concentrated. The preferences of European and Asian countries are more spread out, more on the negative and sad side, and tend to blend into each other rather than form two distinct clusters. One reason for why the listening habits of European and Asian countries are more similar is their location and that countries in Eastern Asia and Western Europe have more chances to share their culture and other factors that lead to more homogenous habits. However, the European and Asian countries are not as concentrated as the America cluster, which could be because the countries in these two continents didn't completely adapt the song preferences of the neighboring countries, but integrated them into their own listening habits. If we look at the scatter plots with all of the continents (including Oceania and Africa), we see that the scatter plot points tend to fall in the middle of the graph by the European countries. One reason for why this could be true is that European tastes and culture were transported to these African and Oceanic countries, thus molding their music tastes and preferences to their European counterparts.

## Limitations

One of our limitations is that while we had a relationship between government data and song positivity, this doesn't necessarily mean that there is a causation between these two variables. Even though we see that governing bodies with democracies have higher song positivity, we can't conclude that democratic countries cause a countries' music preferences to increase.

Another limitation is that we were only looking at the top $50$ songs for each country so we might not have gotten a comprehensive picture of a country's listening preferences. While a vast majority of a country probably does listen to a certain number of songs on the country's top $50$ list, there are probably large groups of people in countries who do not listen to any of the top $50$ songs and have different musical tastes. Thus, our analysis wouldn't apply to these people since they do not listen to any of the songs that we were looking at in our project.

Another limitation is that the dataset we used with our government trust information was created in $2014$ which is a bit outdated. Since government institutions change with every election and as public events unfold, the government trust values in $2014$ might not be the same as $2021$ with different political leaders or with different public events happening (i.e., the pandemic might've decreased citizens trust in their governments ability to act in their best interest). 

One final limitation is that, as aforementioned, not every African or Middle Eastern country was sampled while nearly every American, European, and Asian country was sampled. This lack of data may be because more developing countries are centered in those continents, or because the dataset provider deemed that Spotify was not widely used enough to warrant statistics for those countries. However, it is worth mentioning that in the countries where the data was sampled, the data was collected in a randomized manner--in rural and urban areas, utilizing both phone and face-to-face methods to reach all types of citizens.
